**1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.**


Математическое ожидание и дисперсия случайной непрерывной величины, следующей непрерывному равномерному распределению на отрезке $[a, b]$ описываются формулами: 

$$M(A) = \dfrac{a + b}{2}$$

$$D(A) = \dfrac{(b - a)^2}{12}$$

При решении задачи, несмотря на то, что точка $200$ выколота, мы все равно ее включаем.

Тогда $M(A) = \dfrac{200 + 800}{2} = 500, D(A) = \dfrac{(800 - 200)^2}{12} = 30000$

**2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.**


Известно, что $D(B) = \dfrac{(b - a)^2}{12} = 0.2, a = 0.5$, тогда: $$0.2 = \dfrac{(b - 0.5)^2}{12}$$
$$2.4 = b^2 - b + 0.25$$
$$b^2 - b - 2.15 = 0, $$решив квадратное уравнение, получаем:
$b = \frac{1 \pm \sqrt{9.6}}{2}\quad$ $b_1 = 2.049 \quad b_2 = −1.049$

Так как мы ищем правую границу, то остается единственное значение для правой границы: $b_1 = 2.049$

Тогда $M(B) = \dfrac{a + b}{2} = \dfrac{0.5 + 2.049}{2} = 1.2745$

**3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения
$f(x) = \dfrac{1}{4\sqrt{2\pi}} \cdot \exp \left (- {\frac{(x+2)^2}{32}} \right)$.**
Найдите:

а). $M(X)$

б). $D(X)$

в). $std(X)$ (среднее квадратичное отклонение)


Плотность распределения непрерывной величины $X$, следующей нормальному распределению, задается формулой: $$f(x) = \dfrac{1}{\sqrt{2 \pi \sigma^2}} \cdot \exp \left(- \frac{(x - \mu)^2}{2 \sigma^2} \right)$$

Параметры $\mu$ и $\sigma$ задают, соответственно, математическое ожидание и среднее квадратическое отклонение случайной величины:
$$M(X) = \mu, \:\: D(X) = \sigma^2$$

а) $M(X) = -2$

б) $D(X) = 16$

в) $std(X) = \sigma = 4$

4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.


Вероятность попасть в интервал $(a, b)$ можно посчитать:
$$P(a < X < b) = F(b) - F(a)$$

Для того, чтобы воспользоваться таблицей распределений (где указаны вероятности величины $X$, когда известно на сколько сигм $σ$ отклоняется данная величина от математического ожидания) необходимо нормировать значения $a$ и $b$, приведя нормальное распределение к стандартному нормальному распределению $N(\mu=0, \sigma=1)$ (как StandardScaler() из sklearn). При этом $μ$ и $σ$ нам известны по условиям задачи. Тогда искомая величина выражается формулой: $$P(a < X < b) = \Phi \left(\frac{b - \mu}{\sigma}\right) - \Phi \left(\frac{a - \mu}{\sigma} \right)$$ 

a) $P(X > 182) = \Phi \left(\frac{+\infty - 174}{8}\right) - \Phi \left(\frac{182 - 174}{8} \right) = 0.5000 - 0.3413 = 0.1587$

б) $P(X > 190) = \Phi \left(\frac{+\infty - 174}{8}\right) - \Phi \left(\frac{190 - 174}{8} \right) = 0.5000 - 0.4772 = 0.0228$

в) $P(166 < X < 190) = \Phi \left(\frac{190 - 174}{8}\right) - \Phi \left(\frac{166 - 174}{8} \right) = 0.4772 - (-0.3413) = 0.8185$

г) $P(166 < X < 182) = \Phi \left(\frac{182 - 174}{8}\right) - \Phi \left(\frac{166 - 174}{8} \right) = 0.3413 - (-0.3413) = 0.6826$ (отклонение на +/- $σ$)

д) $P(158 < X < 190) = \Phi \left(\frac{190 - 174}{8}\right) - \Phi \left(\frac{158 - 174}{8} \right) = 0.4772 - (-0.4772) = 0.9544$ (отклонение на +/- $2σ$)

е) $P(X < 150 | X > 190) = \Phi \left(\frac{150 - 174}{8}\right) - \Phi \left(\frac{-\infty - 174}{8} \right) + \Phi \left(\frac{+\infty - 174}{8}\right) - \Phi \left(\frac{190 - 174}{8} \right) = -0.4987 -(-0.5000) + 0.5000 - 0.4772 = 0.0241$

ё) $P(X < 150 | X > 198) = \Phi \left(\frac{150 - 174}{8}\right) - \Phi \left(\frac{-\infty - 174}{8} \right) + \Phi \left(\frac{+\infty - 174}{8}\right) - \Phi \left(\frac{198 - 174}{8} \right) = -0.4987 -(-0.5000) + 0.5000 - 0.4987 = 0.0026$

ж) $P(X < 166) = \Phi \left(\frac{166 - 174}{8}\right) - \Phi \left(\frac{-\infty - 174}{8} \right) = -0.3413 -(-0.5000) = 0.1587$

**5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?**


$Z = \dfrac{190 - M(X)}{\sqrt{D(X)}} = \dfrac{190 - M(X)}{\sigma} = \dfrac{190 - 178}{5} = 2.4$
